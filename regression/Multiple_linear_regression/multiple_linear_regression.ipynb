# Multiple linear regression

A multiple linear regression model involves more than one independent variable and a single dependent variable, and it fits a hyperplane to the data.

**Example of multiple linear regression:** Let’s say we have data on the sales of a company’s products. We have information on the number of advertisements (in thousands) made on TV, radio, and newspapers, as well as the sales figures (in thousands of dollars). Our goal is to build a multiple linear regression model to predict sales based on the advertising budget. The data used in this example can be downloaded from this [link](https://gist.github.com/SoumenAtta/176e6b743778db47dddcf12e476de9a3).

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
# Load the data
df = pd.read_csv("advertising.csv")
df.head()
```

## We split the data into features (X) and target (y). This can be done using the following command:

```python
# Split the data into features (X) and target (y)
X = df[["TV", "Radio", "Newspaper"]].values
y = df["Sales"].values
```

## We are now ready to fit the multiple linear regression model to the input data.

```python
# Fit the multiple linear regression model to the data
reg = LinearRegression().fit(X, y)
```

## We can now make a prediction for a new data point [230.1, 37.8, 69.2].

```python
# Make a prediction for a new data point
X_new = np.array([[230.1, 37.8, 69.2]])
sales_prediction = reg.predict(X_new)
```

# The prediction can be printed as follows:

```python
# Print the prediction
print("The predicted sales are: ${:.2f}k".format(sales_prediction[0]))
```

## Finally, we can plot the data for each of the advertising budgets. This can be done as follows:

## TV Advertising vs Sales

```python
# Plot TV Advertising vs Sales
plt.scatter(df["TV"], df["Sales"])
plt.xlabel("TV Advertising (in thousands)")
plt.ylabel("Sales (in thousands)")
plt.title("TV Advertising vs Sales")
plt.show()
```

## Radio Advertising vs Sales

```python
plt.scatter(df["Radio"], df["Sales"])
plt.xlabel("Radio Advertising (in thousands)")
plt.ylabel("Sales (in thousands)")
plt.title("Radio Advertising vs Sales")
plt.show()
```

## Newspaper Advertising vs Sales

```python
plt.scatter(df["Newspaper"], df["Sales"])
plt.xlabel("Newspaper Advertising (in thousands)")
plt.ylabel("Sales (in thousands)")
plt.title("Newspaper Advertising vs Sales")
plt.show()
```

**Multiple Linear Regression Model**

**Explanation of the entire program:** The first part of the code loads the advertising data into a pandas DataFrame and splits the data into features (X) and target (y). The features are the advertising budgets for TV, radio, and newspaper, and the target is the sales.

The `LinearRegression().fit(X, y)` line fits the multiple linear regression model to the data. The `fit` function takes the features (advertising budgets) and target (sales) as input and returns a regression model that has been trained on the input data.

The next part of the code uses the `reg.predict(X_new)` function to make a prediction for a new advertising campaign with budgets of 230.1k for TV, 37.8k for radio, and 69.2k for newspaper. The argument to the `predict` function is a 2D array, with one row and three columns, representing the advertising budgets. The result is an array with one element, which is the predicted sales.

Finally, the code uses `matplotlib` to plot the data for each of the advertising budgets. The `plt.scatter` function plots the data as a scatter plot.
