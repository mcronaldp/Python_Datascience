```python
# https://www.youtube.com/watch?v=DrjIexCTF70
from datetime import datetime, timedelta
from azure.storage.blob import BlobServiceClient, generate_blob_sas, BlobSasPermissions
import pandas as pd
```

# Credentials

```python
account_name = 'bidevls'
account_key = '' #acces key als je kijkt in Azure onder datalake aan linker kant.
container_name = 'snapshots'
```

# Create connectiestring

```python
connect_str = 'DefaultEndpointsProtocol=Https;AccountName=' + account_name + ';AccountKey=' + account_key + ';EndpointSuffix=core.windows.net'
blob_service_client = BlobServiceClient.from_connection_string(connect_str)
```

# Use client to connect to container

```python
container_client = blob_service_client.get_container_client(container_name)
```

# Get list of all blob files in the container

```python
blob_list = []
for blob_i in container_client.list_blobs():
    blob_list.append(blob_i.name)
```
