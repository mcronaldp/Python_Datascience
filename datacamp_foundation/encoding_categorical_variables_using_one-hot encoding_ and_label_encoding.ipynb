# Handle categorical variables:

```python
import pandas as pd 
# Load the dataset from the Excel file
data = pd.read_excel(r'files\Data_atrition_dataset.xlsx')
data.head()
```

## Identificeren van categorical variables
### Dit zijn Gender, education level, job role en Attrition
### We printen even de datatypes van deze velden

```python
categorical_vars = ['Gender', 'Education Level', 'Job Role', 'Attrition'] 
# Confirm the data types
print(data[categorical_vars].dtypes)
```

# One Hot encoding

#### One-hot encoding transforms each categorical feature into multiple binary features, where each binary feature represents a unique category. You can use the get_dummies function from pandas to perform one-hot encoding: df = df.dummies()

```python
# Perform one-hot encoding
data_encoded = pd.get_dummies(data, columns=categorical_vars)
```

```python
data_encoded.head()
```

## Label encoding
### Label encoding zorgt er voor dat elke unieke waarde (Distinct) een eigen nummer krijgt. 
#### Job role heeft 4 nummers, want er zijn 4 verschillende job roles.

```python
# Perform label encoding (alternative): Label encoding assigns a unique numerical label to each category. You can use the LabelEncoder class from scikit-learn to perform label encoding:
from sklearn.preprocessing import LabelEncoder 
# Perform label encoding
label_encoder = LabelEncoder()
for var in categorical_vars:
    data[var] = label_encoder.fit_transform(data[var])
```

```python
data.head()
```

## Save One Hot encoded dataset:
You can save the encoded dataset to a new Excel file using the to_excel method of the DataFrame:

```python
# Save the encoded dataset to a new Excel file, Dit is de data van One-Hot-Encoding
data_encoded.to_excel(r'files\encoded_dataset.xlsx', index=False)
```
